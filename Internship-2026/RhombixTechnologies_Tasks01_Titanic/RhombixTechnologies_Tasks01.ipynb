- Did **age** matter — were children saved first?

### ❓ Missing Values

Missing values must be handled before training — models cannot work with `NaN`.
We check **count** and **percentage** of missing values per column.

```python
# ─────────────────────────────────────────────
# MISSING VALUES — count and percentage
# ─────────────────────────────────────────────
def missing_summary(df, name):
    # Count nulls per column
    missing_count = df.isnull().sum()
    # Calculate percentage
    missing_pct   = (missing_count / len(df)) * 100
    # Combine into a summary DataFrame
    summary = pd.DataFrame({
        'Missing Count': missing_count,
        'Missing %': missing_pct.round(2)
    })
    # Only show columns that actually have missing values
    summary = summary[summary['Missing Count'] > 0].sort_values('Missing %', ascending=False)
    print(f'=== {name} — Missing Values ===')
    print(summary)
    print()

missing_summary(train, 'TRAIN')
missing_summary(test,  'TEST')
```

### 🎯 Target Variable Distribution

How many passengers survived vs did not survive?

### 3.1 🚻 Survival by Gender

The Titanic followed **'women and children first'** protocol.
We expect females to have a much higher survival rate than males.

```python
# ─────────────────────────────────────────────
# SURVIVAL BY GENDER
# ─────────────────────────────────────────────

# Count and rate of survival grouped by Sex
gender_survival = train.groupby('Sex')['Survived'].agg(['sum', 'count', 'mean'])
gender_survival.columns = ['Survived', 'Total', 'Survival Rate']
gender_survival['Survival Rate'] = (gender_survival['Survival Rate'] * 100).round(2)
print('=== Survival by Gender ===')
print(gender_survival)
print()

# Plot: count plot split by gender and survival
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: count of survived vs not by gender
sns.countplot(data=train, x='Sex', hue='Survived',
              palette={0: '#e74c3c', 1: '#2ecc71'}, ax=axes[0])
axes[0].set_title('Survival Count by Gender', fontweight='bold')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Count')
axes[0].legend(['Did Not Survive', 'Survived'])

# Right: survival rate (%) by gender
gender_survival['Survival Rate'].plot(kind='bar', color=['#3498db', '#e91e8c'],
                                       edgecolor='black', ax=axes[1])
axes[1].set_title('Survival Rate (%) by Gender', fontweight='bold')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Survival Rate (%)')
axes[1].set_xticklabels(['Female', 'Male'], rotation=0)
# Annotate bars
for p in axes[1].patches:
    axes[1].annotate(f'{p.get_height():.1f}%',
                     (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                     ha='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
print('📌 Females had ~74% survival rate vs only ~19% for males.')
```

### 3.2 🎫 Survival by Passenger Class (Pclass)

Pclass represents **socio-economic status**:
- **1st class** = wealthy (upper deck, closer to lifeboats)
- **2nd class** = middle class
- **3rd class** = poor (lower decks, harder to escape)

```python
# ─────────────────────────────────────────────
# SURVIVAL BY PASSENGER CLASS
# ─────────────────────────────────────────────

# Survival stats grouped by Pclass
pclass_survival = train.groupby('Pclass')['Survived'].agg(['sum', 'count', 'mean'])
pclass_survival.columns = ['Survived', 'Total', 'Survival Rate']
pclass_survival['Survival Rate'] = (pclass_survival['Survival Rate'] * 100).round(2)
print('=== Survival by Pclass ===')
print(pclass_survival)
print()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: count by class and survival
sns.countplot(data=train, x='Pclass', hue='Survived',
              palette={0: '#e74c3c', 1: '#2ecc71'}, ax=axes[0])
axes[0].set_title('Survival Count by Pclass', fontweight='bold')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Count')
axes[0].legend(['Did Not Survive', 'Survived'])

# Right: survival rate by class
pclass_survival['Survival Rate'].plot(kind='bar', color=['#f1c40f', '#e67e22', '#e74c3c'],
                                       edgecolor='black', ax=axes[1])
axes[1].set_title('Survival Rate (%) by Pclass', fontweight='bold')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Survival Rate (%)')
axes[1].set_xticklabels(['1st Class', '2nd Class', '3rd Class'], rotation=0)
for p in axes[1].patches:
    axes[1].annotate(f'{p.get_height():.1f}%',
                     (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                     ha='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
print('📌 1st class: ~63% survival | 2nd class: ~47% | 3rd class: ~24%')
```

### 3.3 👶 Survival by Age

Age has **177 missing values** in train (~20%).
We will analyze the age distribution and see if younger passengers survived more.

```python
# ─────────────────────────────────────────────
# SURVIVAL BY AGE
# ─────────────────────────────────────────────

# Basic age stats split by survival
print('=== Age Stats by Survival ===')
print(train.groupby('Survived')['Age'].describe().round(2))
print()

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Left: age distribution for survivors vs non-survivors
# KDE (density) plot shows the shape of age distribution
train[train['Survived'] == 0]['Age'].dropna().plot(kind='kde', ax=axes[0],
    color='#e74c3c', label='Did Not Survive', linewidth=2)
train[train['Survived'] == 1]['Age'].dropna().plot(kind='kde', ax=axes[0],
    color='#2ecc71', label='Survived', linewidth=2)
axes[0].set_title('Age Distribution by Survival', fontweight='bold')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Density')
axes[0].legend()

# Right: boxplot of age by survival
sns.boxplot(data=train, x='Survived', y='Age',
            palette={0: '#e74c3c', 1: '#2ecc71'}, ax=axes[1])
axes[1].set_title('Age Boxplot by Survival', fontweight='bold')
axes[1].set_xlabel('Survived (0=No, 1=Yes)')
axes[1].set_ylabel('Age')

plt.tight_layout()
plt.show()
print('📌 Children (age < 10) had higher survival. Median age of survivors is slightly lower.')
```

### 3.4 💰 Survival by Fare

Fare is closely linked to Pclass — higher fare = higher class = better survival odds.

```python
# ─────────────────────────────────────────────
# SURVIVAL BY FARE
# ─────────────────────────────────────────────

print('=== Fare Stats by Survival ===')
print(train.groupby('Survived')['Fare'].describe().round(2))
print()

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Left: fare distribution (log scale to handle extreme outliers)
train[train['Survived'] == 0]['Fare'].plot(kind='kde', ax=axes[0],
    color='#e74c3c', label='Did Not Survive', linewidth=2)
train[train['Survived'] == 1]['Fare'].plot(kind='kde', ax=axes[0],
    color='#2ecc71', label='Survived', linewidth=2)
axes[0].set_title('Fare Distribution by Survival', fontweight='bold')
axes[0].set_xlabel('Fare')
axes[0].set_ylabel('Density')
axes[0].set_xlim(0, 300)   # Limit x-axis to remove extreme outliers from view
axes[0].legend()

# Right: boxplot of fare by survival
sns.boxplot(data=train, x='Survived', y='Fare',
            palette={0: '#e74c3c', 1: '#2ecc71'}, ax=axes[1])
axes[1].set_title('Fare Boxplot by Survival', fontweight='bold')
axes[1].set_xlabel('Survived (0=No, 1=Yes)')
axes[1].set_ylabel('Fare')
axes[1].set_ylim(0, 300)   # Limit y-axis for better readability

plt.tight_layout()
plt.show()
print('📌 Survivors paid significantly higher fares on average — confirms class advantage.')
```

### 3.5 ⚓ Survival by Embarkation Port

- **C** = Cherbourg (France)
- **Q** = Queenstown (Ireland)
- **S** = Southampton (England) — most passengers boarded here

```python
# ─────────────────────────────────────────────
# SURVIVAL BY EMBARKED PORT
# ─────────────────────────────────────────────

embarked_survival = train.groupby('Embarked')['Survived'].agg(['sum', 'count', 'mean'])
embarked_survival.columns = ['Survived', 'Total', 'Survival Rate']
embarked_survival['Survival Rate'] = (embarked_survival['Survival Rate'] * 100).round(2)
print('=== Survival by Embarked Port ===')
print(embarked_survival)
print()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: count by port and survival
sns.countplot(data=train, x='Embarked', hue='Survived',
              palette={0: '#e74c3c', 1: '#2ecc71'}, ax=axes[0])
axes[0].set_title('Survival Count by Embarked Port', fontweight='bold')
axes[0].set_xlabel('Embarked Port (C=Cherbourg, Q=Queenstown, S=Southampton)')
axes[0].set_ylabel('Count')
axes[0].legend(['Did Not Survive', 'Survived'])

# Right: survival rate by port
embarked_survival['Survival Rate'].plot(kind='bar', color=['#9b59b6', '#1abc9c', '#e67e22'],
                                         edgecolor='black', ax=axes[1])
axes[1].set_title('Survival Rate (%) by Embarked Port', fontweight='bold')
axes[1].set_xlabel('Embarked Port')
axes[1].set_ylabel('Survival Rate (%)')
axes[1].set_xticklabels(['Cherbourg (C)', 'Queenstown (Q)', 'Southampton (S)'], rotation=15)
for p in axes[1].patches:
    axes[1].annotate(f'{p.get_height():.1f}%',
                     (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                     ha='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
print('📌 Cherbourg passengers had highest survival (~55%) — likely more 1st class passengers.')
```

### 3.6 👨‍👩‍👧 Survival by Family Size

We combine `SibSp` (siblings/spouses) + `Parch` (parents/children) to get total **FamilySize**.
- **Alone** (FamilySize = 1) — travelling solo
- **Small family** (2–4) — may have helped each other
- **Large family** (5+) — harder to coordinate escape

```python
# ─────────────────────────────────────────────
# SURVIVAL BY FAMILY SIZE
# ─────────────────────────────────────────────

# Create FamilySize = SibSp + Parch + 1 (include the passenger themselves)
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1

family_survival = train.groupby('FamilySize')['Survived'].agg(['sum', 'count', 'mean'])
family_survival.columns = ['Survived', 'Total', 'Survival Rate']
family_survival['Survival Rate'] = (family_survival['Survival Rate'] * 100).round(2)
print('=== Survival by Family Size ===')
print(family_survival)
print()

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Left: count by family size and survival
sns.countplot(data=train, x='FamilySize', hue='Survived',
              palette={0: '#e74c3c', 1: '#2ecc71'}, ax=axes[0])
axes[0].set_title('Survival Count by Family Size', fontweight='bold')
axes[0].set_xlabel('Family Size (1 = Alone)')
axes[0].set_ylabel('Count')
axes[0].legend(['Did Not Survive', 'Survived'])

# Right: survival rate by family size
family_survival['Survival Rate'].plot(kind='bar', edgecolor='black',
                                       color='#3498db', ax=axes[1])
axes[1].set_title('Survival Rate (%) by Family Size', fontweight='bold')
axes[1].set_xlabel('Family Size')
axes[1].set_ylabel('Survival Rate (%)')
for p in axes[1].patches:
    axes[1].annotate(f'{p.get_height():.1f}%',
                     (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                     ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
print('📌 Small families (2-4) had best survival. Solo travellers and large families did worse.')
```

### 3.7 🌡️ Correlation Heatmap

A **correlation heatmap** shows how strongly each numeric feature is related to every other feature.
- Values close to **+1** = strong positive correlation
- Values close to **-1** = strong negative correlation
- Values close to **0** = no correlation

We focus on which features correlate most with `Survived`.

```python
# ─────────────────────────────────────────────
# CORRELATION HEATMAP
# ─────────────────────────────────────────────

# Select only numeric columns for correlation
numeric_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'FamilySize']
corr_matrix = train[numeric_cols].corr()

# Print correlation with Survived specifically
print('=== Correlation with Survived ===')
print(corr_matrix['Survived'].sort_values(ascending=False).round(3))
print()

# Plot full heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,          # Show correlation values inside cells
    fmt='.2f',           # 2 decimal places
    cmap='coolwarm',     # Red = positive, Blue = negative
    center=0,            # Center colormap at 0
    linewidths=0.5,      # Grid lines between cells
    square=True          # Square cells
)
plt.title('Correlation Heatmap of Numeric Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
print('📌 Fare has strongest positive correlation with Survived (+0.26).')
print('📌 Pclass has strongest negative correlation with Survived (-0.34) — lower class = less survival.')
```
